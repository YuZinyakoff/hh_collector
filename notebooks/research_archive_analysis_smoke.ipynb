# Research Archive Analysis Smoke

This notebook proves that a small Research Archive v1 sample can be downloaded/read into a DataFrame and plotted. It is not production analysis; the current VPS corpus is treated as `pilot/test corpus` evidence.

```python
from pathlib import Path

ARCHIVE_DIR = Path('.state/analysis/research-archive-s3-sample')
OUTPUT_DIR = Path('.state/analysis/research-archive-smoke')
ARCHIVE_DIR, OUTPUT_DIR
```

Download a bounded sample from configured S3. Requires `.env` with `HHRU_RESEARCH_ARCHIVE_OFFSITE_S3_*` or fallback `HHRU_BACKUP_OFFSITE_S3_*` values.

```python
!PYTHONPATH=src python scripts/analysis/research_archive_dataframe_smoke.py \
  --archive-dir {ARCHIVE_DIR} \
  --output-dir {OUTPUT_DIR} \
  --download-from-s3 \
  --max-manifests 5 \
  --max-rows 5000
```

```python
import pandas as pd

df = pd.read_csv(OUTPUT_DIR / 'sample_rows.csv')
df.head()
```

```python
df['_archive_dataset_key'].value_counts().sort_index().plot.bar(title='Rows by dataset')
```
